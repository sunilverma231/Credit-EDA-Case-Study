## Missing Data and Cleaning 
### Search for null values and check how many of the columns have null values more than 40% and Delete such columns 

```python
#check the shape of columns before checing it 
appdata[appdata['EXT_SOURCE_1'].isna() & appdata['EXT_SOURCE_2'].isna() & appdata['EXT_SOURCE_3'].isna()].shape
```

## Binning Data

```python
#Binning data from AMT_INCOME_TOTAL in to different groups on basis of quantiles 
q1=appdata['AMT_INCOME_TOTAL'].quantile(0.25)
q2=appdata['AMT_INCOME_TOTAL'].quantile(0.50)
q3=appdata['AMT_INCOME_TOTAL'].quantile(0.75)
m=appdata['AMT_INCOME_TOTAL'].max()

# Binning AMT_INCOME_TOTAL into AMT_INCOME_TOTAL_bin so we don't loose data and have binned values
appdata['AMT_INCOME_TOTAL_bin'] = pd.cut(appdata['AMT_INCOME_TOTAL'],[q1, q2, q3,m ], labels = ['Low', 'medium', 'High'])
print(appdata.AMT_INCOME_TOTAL_bin.value_counts())
```

```python
# Binning data from AMT_CREDIT in to different groups on basis of quantiles 
# AMT_CREDIT
q1=appdata['AMT_CREDIT'].quantile(0.25)
q2=appdata['AMT_CREDIT'].quantile(0.50)
q3=appdata['AMT_CREDIT'].quantile(0.75)
m=appdata['AMT_CREDIT'].max()

# Binning AMT_CREDIT into AMT_CREDIT_bin so we don't loose data and have binned values
appdata['AMT_CREDIT_bin'] = pd.cut(appdata['AMT_CREDIT'],[q1, q2, q3,m ], labels = ['Low', 'medium', 'High'])
print(appdata.AMT_CREDIT_bin.value_counts())
```
